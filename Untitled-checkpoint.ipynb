```python
%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import numpy as np
ax = plt.gca()
r = Rectangle((0.1,0.1), 0.5, 0.5, fill=False, ec="r")
ax.add_patch(r)
xl,xu = 0.1, 0.6
yl,yu = 0.1, 0.6
plt.gcf().canvas.mpl_connect('button_press_event', click)

def click(event):

    x= event.xdata
    y= event.ydata
    #check if click is inside axes
    if event.inaxes:
        if (xl <=x <= xu and yl<=y <=yu):
            plt.text(x, y, 'o')
        else: 
            plt.text(x, y, 'x')
plt.draw()
plt.show()
plt.savefig("Prob 1Fig")
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

mean = 2
sd = 0.5
# Generate list of 1000 sample from normal distribution with mean=2 and sd=0.5
samples = np.random.normal(mean,sd,1000)

# Create a plot with bins=50 and histtype=step
plt.hist(samples, bins=50 ,histtype='step')
# Rotate X ticks with 45 degrees
plt.xticks(rotation= 45)
# Add y label
plt.ylabel("Frequency")
#Add plot title
plt.title('Lengths of Dogs in Greensboro in feet')
#Save the plot to a file called plot.png
plt.savefig('plot.png')
#Display the plot inline
plt.show()
```
